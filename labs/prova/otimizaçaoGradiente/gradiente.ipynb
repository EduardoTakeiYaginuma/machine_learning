# Desafios de Métodos de Otimização em Altas Dimensões

Métodos de otimização que funcionam bem em pequenas dimensões podem falhar em altas dimensões devido a vários fatores geométricos e estatísticos:

- **Curse of Dimensionality**: À medida que a dimensão aumenta, o volume do espaço cresce exponencialmente, fazendo com que os dados fiquem mais esparsos. Isso pode dificultar a generalização dos modelos e levar a soluções não ótimas.

- **Distâncias Relativas**: Em alta dimensionalidade, os pontos se tornam equidistantes entre si. Isso significa que métodos que dependem de distâncias (como gradientes ou medidas de similaridade) podem se tornar ineficazes, pois as diferenças entre os pontos se tornam menos significativas.

- **Superfícies de Otimização Complexas**: Em dimensões elevadas, as superfícies de perda se tornam mais complexas, com múltiplos mínimos locais. Métodos que funcionam em dimensões baixas podem ficar presos em mínimos locais sem encontrar o mínimo global.

- **Sensibilidade a Perturbações**: Pequenas mudanças em parâmetros podem levar a grandes variações no resultado em alta dimensão, tornando a otimização instável.

- **Visualização e Intuição**: É difícil visualizar e entender o comportamento de funções em altas dimensões, o que complica a análise e a escolha de estratégias de otimização adequadas.


# Encontrando Mínimos de Funções Multivariadas

## 1. Encontrando Mínimos Analiticamente

- **Derivadas Parciais**: Para encontrar os mínimos de uma função \( f(\bm{x}) \), você começa calculando as derivadas parciais em relação a cada variável. O resultado é chamado de **gradiente**:

  \[
  \nabla f(\bm{x}) = 
  \begin{bmatrix}
  \frac{\partial f}{\partial x_{1}}(\bm{x}) \\
  \frac{\partial f}{\partial x_{2}}(\bm{x}) \\
  \vdots \\
  \frac{\partial f}{\partial x_{n}}(\bm{x})
  \end{bmatrix}
  \]

- **Igualando a Zero**: Para encontrar os pontos críticos, você define o gradiente igual ao vetor nulo (\(\overrightarrow{\bm{0}}\)). As soluções dessa equação fornecem os **pontos críticos** ou **pontos estacionários**, que podem ser mínimos, máximos ou pontos de sela.

## 2. Classificando os Pontos Críticos

- **Hessiana**: Para determinar a natureza dos pontos críticos, você calcula a **matriz Hessiana**, que contém as segundas derivadas:

  \[
  \bm{H}_{f}(\bm{x}) = 
  \begin{bmatrix}
  \frac{\partial^{2} f}{\partial x_{1}^{2}}(\bm{x}) & \frac{\partial^{2} f}{\partial x_{1} \partial x_{2}}(\bm{x}) & \cdots \\
  \frac{\partial^{2} f}{\partial x_{2} \partial x_{1}}(\bm{x}) & \frac{\partial^{2} f}{\partial x_{2}^{2}}(\bm{x}) & \cdots \\
  \vdots & \vdots & \ddots
  \end{bmatrix}
  \]

- **Autovalores**: Você calcula os autovalores da Hessiana. A interpretação é:
  - Todos positivos: ponto mínimo.
  - Todos negativos: ponto máximo.
  - Mistura de sinais (alguns positivos e negativos): ponto de sela.

## 3. Relação com Machine Learning

- **Uso Limitado de Derivadas Analíticas**: Poucos métodos de machine learning utilizam a derivada analítica para encontrar parâmetros ótimos. O mais comum é a **regressão linear**.

- **Métodos Iterativos**: Muitos métodos usam o gradiente analiticamente derivado, mas não para resolver a equação diretamente. Em vez disso, eles utilizam um método iterativo como o **gradiente descendente**, que ajusta os parâmetros em direção ao mínimo usando o gradiente.

- **Diferenciação Automática**: Uma técnica chamada **diferenciação automática** (ou *autograd*) permite calcular o gradiente numericamente sem precisar da fórmula. Isso é crucial em deep learning, onde bibliotecas como TensorFlow, PyTorch e JAX implementam essa técnica, facilitando a otimização de funções de perda em relação a parâmetros ajustáveis.

## Resumo

Em essência, o trecho explica um método formal para encontrar mínimos de funções multivariadas usando derivadas, além de relacionar esses conceitos à prática em machine learning, onde métodos iterativos e diferenciação automática são predominantes.


# Resumo Teórico da Descida por Gradiente

A descida por gradiente é um algoritmo de otimização amplamente utilizado para encontrar o mínimo de funções, especialmente em problemas de aprendizado de máquina e estatística. O princípio fundamental é que, para minimizar uma função \( f(\mathbf{x}) \), você deve se mover na direção oposta ao gradiente da função em relação aos parâmetros.

## Conceitos Fundamentais

- **Gradiente**: O gradiente de uma função \( f \) é um vetor que contém as derivadas parciais de \( f \) em relação a todas as suas variáveis. Ele indica a direção de maior aumento da função. Assim, para minimizar \( f \), movemos na direção oposta ao gradiente.

  \[
  \nabla f(\mathbf{x}) = 
  \begin{bmatrix}
  \frac{\partial f}{\partial x_1} \\[10pt]
  \frac{\partial f}{\partial x_2} \\[10pt]
  \vdots \\[10pt]
  \frac{\partial f}{\partial x_n}
  \end{bmatrix}
  \]

- **Atualização dos Parâmetros**: A atualização dos parâmetros \( \mathbf{x} \) é feita da seguinte forma:

  \[
  \mathbf{x}_{\text{new}} = \mathbf{x}_{\text{old}} - \eta \nabla f(\mathbf{x}_{\text{old}})
  \]

  Onde \( \eta \) é a **taxa de aprendizado**, um hiperparâmetro que controla a magnitude do passo.

- **Condições de Parada**: O algoritmo continua a iterar até que o gradiente seja suficientemente pequeno (indicando que se aproximou de um mínimo) ou até que um número máximo de iterações seja alcançado.

## Propriedades e Desafios

- **Convergência**: A descida por gradiente geralmente converge para um mínimo local. Em funções não convexas, pode não encontrar o mínimo global. A escolha do ponto inicial pode impactar o resultado final.

- **Taxa de Aprendizado**: Uma taxa de aprendizado muito alta pode causar oscilações ou divergência, enquanto uma taxa muito baixa pode levar a uma convergência muito lenta. Métodos como a descida por gradiente estocástico (SGD) ou o uso de técnicas adaptativas (como Adam) podem ajudar a otimizar essa escolha.

- **Múltiplos Mínimos Locais**: Em problemas com múltiplos mínimos, a descida por gradiente pode ser sensível à inicialização. Testar diferentes pontos iniciais pode ser uma estratégia útil.


```python
import numpy as np
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(f, g, x0, eta, tol, max_iter=1000):
    x = x0
    all_x = [x]
    for i in range(max_iter):
        grad = g(x)
        x = x - eta * grad
        all_x.append(x)
        if np.linalg.norm(grad) < tol:
            break
    return x, all_x

# Função com múltiplos mínimos locais
def f(x):
    return np.sin(x) + 0.1 * x**2

def grad_f(x):
    return np.cos(x) + 0.2 * x

# Configuração do plot
plt.figure(figsize=(15, 10))

# Plot da função
x_values = np.linspace(-10, 10, 1000)
y_values = f(x_values)
plt.plot(x_values, y_values, label='f(x) = sin(x) + 0.1x^2', color='blue')

# Diferentes taxas de aprendizado
learning_rates = [0.01, 0.1, 0.5]
colors = ['red', 'green', 'orange']

for eta, color in zip(learning_rates, colors):
    initial_guesses = np.linspace(-10, 10, 5)
    for x0 in initial_guesses:
        x_min, path = gradient_descent(f, grad_f, x0, eta, 1e-6)
        path = np.array(path)
        plt.plot(path, f(path), color=color, alpha=0.5, marker='o', markersize=3)
        plt.annotate(f'η={eta}', (path[-1], f(path[-1])), xytext=(5,5), textcoords='offset points')

plt.title('Gradient Descent Paths for Various Learning Rates and Initial Guesses')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Análise do comportamento
print("Análise do comportamento do algoritmo:")
for eta in learning_rates:
    convergence_points = []
    for x0 in np.linspace(-10, 10, 20):
        x_min, _ = gradient_descent(f, grad_f, x0, eta, 1e-6)
        convergence_points.append(x_min)
    
    unique_minima = np.unique(np.round(convergence_points, 2))
    print(f"\nTaxa de aprendizado η={eta}:")
    print(f"  Número de mínimos encontrados: {len(unique_minima)}")
    print(f"  Mínimos: {unique_minima}")

# Visualização do espaço de parâmetros
plt.figure(figsize=(12, 6))
x = np.linspace(-10, 10, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = f(X)

plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(label='f(x)')

for eta, color in zip(learning_rates, colors):
    x0 = -8
    x_min, path = gradient_descent(f, grad_f, x0, eta, 1e-6)
    path = np.array(path)
    plt.plot(path, [0]*len(path), color=color, marker='o', label=f'η={eta}')

plt.title('Contour Plot of f(x) with Gradient Descent Paths')
plt.xlabel('x')
plt.ylabel('y (apenas para visualização)')
plt.legend()
plt.show()

print("\nInterpretação dos resultados:")
print("1. Mínimo global vs. local:")
print("   - A função f(x) = sin(x) + 0.1x^2 tem múltiplos mínimos locais.")
print("   - Dependendo do ponto inicial e da taxa de aprendizado, o algoritmo pode convergir para diferentes mínimos.")
print("   - O mínimo global está próximo de x ≈ -1.4")

print("\n2. Efeito da taxa de aprendizado:")
print("   - Taxas muito baixas (η=0.01) podem resultar em convergência lenta ou falha em escapar de mínimos locais.")
print("   - Taxas muito altas (η=0.5) podem causar oscilações ou divergência.")
print("   - Uma taxa intermediária (η=0.1) geralmente oferece um bom equilíbrio entre velocidade e estabilidade.")

print("\n3. Importância do ponto inicial:")
print("   - Diferentes pontos iniciais podem levar a diferentes mínimos locais.")
print("   - Em problemas mais complexos, é comum usar múltiplos pontos iniciais ou técnicas como reinício aleatório.")

print("\n4. Critérios de parada:")
print("   - O algoritmo para quando o gradiente está próximo de zero (tolerância) ou após um número máximo de iterações.")
print("   - É importante escolher critérios de parada apropriados para evitar parada prematura ou execução excessiva.")

print("\nConclusão:")
print("A descida do gradiente é uma técnica poderosa, mas requer ajuste cuidadoso dos hiperparâmetros (taxa de aprendizado, ")
print("tolerância, etc.) e consideração da natureza da função objetivo para garantir convergência para o mínimo global.")
```
